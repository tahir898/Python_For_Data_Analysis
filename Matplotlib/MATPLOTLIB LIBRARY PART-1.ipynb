# * creating multiple bar graphs

```python
x=["car","bike","scootie","bicycle"]
y=[40,85,75,35]
z=[20,30,40,15]
c=["yellow"]
plt.title("parking survey")
plt.xlabel("vehicles",fontsize=20)
plt.ylabel("numbers",fontsize=20)
plt.bar(x,y,width=0.5,color=c,edgecolor="blue")
plt.bar(x,z, width=0.5, color="green",edgecolor="blue")
plt.show()
```

```python
import numpy as np
x=["car","bike","scootie","bicycle"]
y=[40,85,75,35]
z=[20,30,40,15]
c=["yellow"]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("parking survey")
plt.xlabel("vehicles",fontsize=20)
plt.ylabel("numbers",fontsize=20)

plt.bar(p,y,width=0.5,color=c,edgecolor="blue")
plt.bar(p1,z, width=0.5, color="green",edgecolor="blue")
plt.show()
```

```python
import numpy as np
x=["car","bike","scootie","bicycle"]
y=[40,85,75,35]
z=[20,30,40,15]
c=["yellow"]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("parking survey")
plt.xlabel("vehicles",fontsize=20)
plt.ylabel("numbers",fontsize=20)

plt.xticks(p+width,x)

plt.bar(p,y,width=0.5,color=c,edgecolor="blue")
plt.bar(p1,z, width=0.5, color="green",edgecolor="blue")
plt.show()
```

```python
import numpy as np
x=["car","bike","scootie","bicycle"]
y=[40,85,75,35]
z=[20,30,40,15]
c=["yellow"]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("parking survey")
plt.xlabel("vehicles",fontsize=20)
plt.ylabel("numbers",fontsize=20)

plt.xticks(p+width/2, x)

plt.bar(p,y,width=0.5,color=c,edgecolor="blue")
plt.bar(p1,z, width=0.5, color="green",edgecolor="blue")
plt.show()
```

```python
import numpy as np
x=["car","bike","scootie","bicycle"]
y=[40,85,75,35]
z=[20,30,40,15]
c=["yellow"]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("parking survey")
plt.xlabel("vehicles",fontsize=20)
plt.ylabel("numbers",fontsize=20)

plt.xticks(p+width/2, x,rotation=30)

plt.bar(p,y,width=0.5,color=c,edgecolor="blue")
plt.bar(p1,z, width=0.5, color="green",edgecolor="blue")
plt.show()
```

```python
import numpy as np
x=["car","bike","scootie","bicycle"]
y=[40,85,75,35]
z=[20,30,40,15]
c=["yellow"]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title("parking survey")
plt.xlabel("vehicles",fontsize=20)
plt.ylabel("numbers",fontsize=20)

plt.xticks(p+width/2, x,rotation=30)

plt.barh(p,y,color=c,edgecolor="blue")
plt.bar(p1,z, color="green",edgecolor="blue")
plt.show()
```

# *changing the size of the dots

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
c=["blue","green","red","yellow","purple","violet"]
plt.scatter(x,y,color=c,s=150)
plt.title("scatter plot",fontsize=20)
plt.xlabel("month",fontsize=20)
plt.ylabel("number",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
c=["blue","green","red","yellow","purple","violet"]
d=[200,300,400,500,600,700]
plt.scatter(x,y,color=c,s=d,marker=".",edgecolor="yellow")
plt.title("scatter plot",fontsize=20)
plt.xlabel("month",fontsize=20)
plt.ylabel("number",fontsize=20)
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
c=["blue","green","red","yellow","purple","violet"]
d=[200,300,400,500,600,700]
plt.scatter(x,y,color=c,s=d,marker=".",edgecolor="yellow")
plt.title("scatter plot",fontsize=20)
plt.colorbar()
plt.xlabel("month",fontsize=20)
plt.ylabel("number",fontsize=20)
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
c=["blue","green","red","yellow","purple","violet"]
d=[200,300,400,500,600,700]
plt.scatter(x,y,color=c,s=d,marker=".",edgecolor="yellow")
plt.title("scatter plot",fontsize=20)
t=plt.colorbar()
t.set_label("colorbar",fontsize=20)
plt.xlabel("month",fontsize=20)
plt.ylabel("number",fontsize=20)
```
