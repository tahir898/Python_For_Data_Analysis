# *whis parameter

```python
x=[10,15,3,40,25,60,70,130]
plt.boxplot(x,widths=0.2,labels=["code"],patch_artist="blue",showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,150]
plt.boxplot(x,widths=0.2,labels=["code"],patch_artist="blue",showmeans=True,sym="g+")
plt.show()
```

# * change the color of the box

```python
x=[10,20,30,40,50,60,70,150]
plt.boxplot(x,widths=0.2,labels=["code"],patch_artist="blue",showmeans=True,boxprops=dict(color="r"))
plt.show()
```

# * changing the color of the cap

```python
x=[10,20,30,40,50,60,70,150]
plt.boxplot(x,widths=0.2,labels=["code"],patch_artist="blue",showmeans=True,boxprops=dict(color="r"),capprops=dict(color="magenta"),whiskerprops=dict(color="aqua"))
plt.show()
```

# *changing the color of the outlier

```python
x=[10,20,30,40,50,60,70,150]
plt.boxplot(x,widths=0.2,labels=["code"],patch_artist="blue",showmeans=True,boxprops=dict(color="r"),capprops=dict(color="magenta"),
            whiskerprops=dict(color="aqua"),flierprops=dict(markeredgecolor="g"))
plt.show()
```

# * multiple box plots

```python
x=[10,20,30,40,50,60,70,150]
x1=[10,20,30,40,50,60,70,90,120]
x2=[25,23,27,28,30,48,50,60,75]

y=[x,x1,x2]

plt.boxplot(y,widths=0.2,labels=["apple","cherry","orange"],patch_artist="blue",showmeans=True,boxprops=dict(color="r"),capprops=dict(color="magenta"),
            whiskerprops=dict(color="aqua"),flierprops=dict(markeredgecolor="g"))
plt.show()
```

# #AREA STACK PLOT

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[1,2,3,4,5]

plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area=[1,2,5,4,1]

plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[1,2,5,4,1]
area2=[1,2,3,4,2]
area3=[4,1,2,3,5]

plt.stackplot(x,area1,area2,area3)
plt.show()
```

# * label

```python
x=[1,2,3,4,5]
area1=[1,2,5,4,1]
area2=[1,2,3,4,2]
area3=[4,1,2,3,5]

l=["area1","area2","area3"]


plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[1,2,5,4,1]
area2=[1,2,3,4,2]
area3=[4,1,2,3,5]

l=["area1","area2","area3"]


plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[1,2,5,4,1]
area2=[1,2,3,4,2]
area3=[4,1,2,3,5]

l=["area1","area2","area3"]


plt.stackplot(x,area1,area2,area3,labels=l,colors=["red","aqua","yellow"])
plt.legend(loc=2)
plt.show()
```

# * base line

```python
x=[1,2,3,4,5]
area1=[1,2,5,4,1]
area2=[1,2,3,4,2]
area3=[4,1,2,3,5]

l=["area1","area2","area3"]


plt.stackplot(x,area1,area2,area3,labels=l,colors=["red","aqua","yellow"],baseline="sym")
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[1,2,5,4,1]
area2=[1,2,3,4,2]
area3=[4,1,2,3,5]

l=["area1","area2","area3"]


plt.stackplot(x,area1,area2,area3,labels=l,colors=["red","aqua","yellow"],baseline="zero")
plt.legend(loc=2)
plt.show()
```

# * title

```python
x=[1,2,3,4,5]
area1=[1,2,5,4,1]
area2=[1,2,3,4,2]
area3=[4,1,2,3,5]

l=["area1","area2","area3"]

plt.title("graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.stackplot(x,area1,area2,area3,labels=l,colors=["red","aqua","yellow"],baseline="zero")
plt.legend(loc=2)
plt.show()
```

# * step plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y)
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y)
plt.grid()
plt.title("graph")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y)
plt.grid()
plt.title("graph",fontsize=20)
plt.xlabel("x_axis",fontsize=20)
plt.ylabel("y_axis",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y,color="r")
plt.grid()
plt.title("graph",fontsize=20)
plt.xlabel("x_axis",fontsize=20)
plt.ylabel("y_axis",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y,color="r",marker="o")
plt.grid()
plt.title("graph",fontsize=20)
plt.xlabel("x_axis",fontsize=20)
plt.ylabel("y_axis",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y,color="r",marker="o",ms=10)
plt.grid()
plt.title("graph",fontsize=20)
plt.xlabel("x_axis",fontsize=20)
plt.ylabel("y_axis",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y,color="r",marker="o",ms=10,mfc="blue")
plt.grid()
plt.title("graph",fontsize=20)
plt.xlabel("x_axis",fontsize=20)
plt.ylabel("y_axis",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]
plt.step(x,y,color="r",marker="o",ms=10,mfc="blue",label="price")
plt.legend()
plt.grid()
plt.title("graph",fontsize=20)
plt.xlabel("x_axis",fontsize=20)
plt.ylabel("y_axis",fontsize=20)
plt.show()
```

# * sub plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x,y,color="red")

x1=[1,2,3,4,5]
y1=[2,3,5,1,4]
plt.step(x,y)

x3=[10,20,30,40]
plt.pie(x3)

x4=["a","b","c","d"]
y4=[2,3,4,5]
plt.bar(x4,y4)

plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="red")

x1=[1,2,3,4,5]
y1=[2,3,5,1,4]
plt.subplot(2,2,2)
plt.step(x1,y1)

x3=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x3)

x4=["a","b","c","d"]
y4=[2,3,4,5]
plt.subplot(2,2,4)
plt.bar(x4,y4)

plt.show()
```

# * how to save figure in matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.plot(x,y)
plt.title("graph")
plt.xlabel("day")
plt.ylabel("number")

plt.savefig("image_03",dpi=200,facecolor="g")

plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.plot(x,y)
plt.title("graph")
plt.xlabel("day")
plt.ylabel("number")

plt.savefig("image_06.jpg",dpi=100,facecolor="g",transparent=True)

plt.show()
```
