```python
x=[10,20,30,40]
y=["english","hindi","maths","science"]
c=["brown","pink","magenta","grey"]

plt.pie(x,labels=y,colors=c,autopct="%0.1f%%")
plt.show()
```

# * shadow

```python
x=[10,20,30,40]
y=["english","hindi","maths","science"]
c=["brown","pink","magenta","grey"]

plt.pie(x,labels=y,colors=c,autopct="%0.1f%%",shadow=True)
plt.show()
```

# * font size

```python
x=[10,20,30,40]
y=["english","hindi","maths","science"]
c=["brown","pink","magenta","grey"]

plt.pie(x,labels=y,colors=c,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.2,
        startangle=90,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":1,"edgecolor":"blue"})
plt.show()
```

# * STEM PLOT

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":",markerfmt="g*")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":",markerfmt="g*",bottom=2)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":",markerfmt="g*",basefmt="g")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":",markerfmt="g*",basefmt="g",label="code")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":",markerfmt="g*",basefmt="g",label="code",use_line_collection=False)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=":",markerfmt="g*",basefmt="g",label="code",use_line_collection=False,orientation="horizontal")
plt.legend()
plt.show()
```
